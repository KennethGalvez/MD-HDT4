## Usando s√©palo


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from kneed import KneeLocator

# Leer el archivo CSV y seleccionar dos columnas espec√≠ficas
data = pd.read_csv('iris.csv', usecols=['sepal_length', 'sepal_width'])

# Convertir los datos a un arreglo de numpy
X = data.values

# Estandarizar los datos
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Calcular la inercia para m√∫ltiples valores de n_clusters
inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Utilizar la funci√≥n KneeLocator para encontrar el codo
kn = KneeLocator(range(1, 11), inertias, curve='convex', direction='decreasing')
n_clusters = kn.knee

# Graficar la curva de la inercia y el codo encontrado
plt.plot(range(1, 11), inertias, marker='o')
plt.axvline(x=n_clusters, linestyle='--', color='r', label='Codo')
plt.xlabel('N√∫mero de Clusters')
plt.ylabel('Inercia')
plt.legend()
plt.show()
```
